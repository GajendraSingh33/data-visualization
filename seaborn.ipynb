```python
# scatter plot using seaborn
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

sns.scatterplot(x='day', y='tip', data=data,)
plt.show()
```

```python
# You will find that while using Matplotlib it will a lot difficult if you want to color each point 
# of this plot according to the sex. But in scatter plot it can be done with the help of hue argument.
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

sns.scatterplot(x='day', y='tip', data=data,
               hue='sex')
plt.show()
```

```python

# line plot in seaborn
# Line Plot in Seaborn plotted using the lineplot() method.  In this, we can pass only the data argument also.
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# reading the database
data = pd.read_csv("tips.csv")

sns.lineplot(x='day', y='tip', data=data)
plt.show()
```

```python
# line plot in seaborn
# example 2
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# reading the database
data = pd.read_csv("tips.csv")

# using only data attribute
sns.lineplot(data=data.drop(['total_bill'], axis=1))
plt.show()
```

```python
# Bar Plot
# Bar Plot in Seaborn can be created using the barplot() method.
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# reading the database
data = pd.read_csv("tips.csv")

sns.barplot(x='day',y='tip', data=data, 
            hue='sex')

plt.show()
```

```python
# histogram in seaborn
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# reading the database
data = pd.read_csv("tips.csv")

sns.histplot(x='total_bill', data=data, kde=True, hue='sex')

plt.show()
```

```python
# Violin Plot in seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your own CSV file
tips = pd.read_csv("tips.csv")

# Create violin plot
sns.violinplot(data=tips, x="day", y="total_bill")

plt.show()
```

```python
# Strip Plot in seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.stripplot(data=tips, x="day", y="total_bill")

plt.show()
```

```python
# Kernel Density Estimate Plot in seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=tips, x="total_bill")

plt.show()
```

```python
# Heatmap in seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your own tips dataset
tips = pd.read_csv("tips.csv")

# Create a pivot table: average total_bill for each (day, time)
tips_pivot = tips.pivot_table(index="day", columns="time", values="total_bill", aggfunc="mean")

# Plot heatmap
sns.heatmap(tips_pivot, annot=True, fmt=".1f", cmap="YlGnBu")

plt.show()
```
